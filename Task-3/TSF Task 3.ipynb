```python
# Count the unique value occurance in Sub-Category using bar chart.

subcat_count = data['Sub-Category'].value_counts()
print(subcat_count)
# Plot a bar chart of the sub-category counts
plt.figure(figsize=(14,8))
subcat_count.plot.bar()
```

#### Insight-
>* Most common sub-category of product is Binders, followed by Paper, Furnishings, Phones, and Storage in descending order.

```python
#Count the number of unique value occurance in 'Quantity'..Quantity of Products Sold by Sub-Category using seaborn library
data.Quantity.value_counts()
subcat_quantity = data.groupby('Sub-Category')['Quantity'].sum().sort_values(ascending=False)
print(subcat_quantity)



# Group quantity by sub-category
subcat_quantity = data.groupby('Sub-Category')['Quantity'].sum().sort_values(ascending=False)

# Create a bar plot using Seaborn
fig, ax = plt.subplots(figsize=(10,8))
sns.barplot(x='Quantity', y='Sub-Category', data=subcat_quantity.reset_index(), ax=ax)
ax.set_title('Quantity of Products Sold by Sub-Category')
ax.set_xlabel('Quantity')
ax.set_ylabel('Sub-Category')
plt.show()
```

#### Insights-
>* The sub-category with the highest quantity of products sold is 'Binders', followed by 'Paper' and 'Furnishings'.
>* 'Copiers', 'Machines', and 'Supplies' have the lowest quantity of products sold.
>* The majority of sub-categories have sold around 1,000-3,000 products. 

```python
#Average sales and average profit by category and region.
plt.figure(figsize=(12,6))

#First Plot
plt.subplot(1, 2, 1)
df_subset = data[['Category', 'Region', 'Sales']]
df_subset = df_subset.groupby(['Category', 'Region']).mean().reset_index()
sns.barplot(x='Category', y='Sales', hue='Region', data=df_subset)
plt.title('Average Sales by Category and Region')

# Second Plot
plt.subplot(1,2,2)
df_subset = data[['Category', 'Region', 'Profit']]
df_subset = df_subset.groupby(['Category', 'Region']).mean().reset_index()
sns.barplot(x='Category', y='Profit', hue='Region', data=df_subset)
plt.title('Average Profit by Category and Region')

# Adjust the layout and display the plots
plt.tight_layout()
plt.show()
```

##### Insights- 
>Ist plot
>>* The first plot indicates that the highest average sales come from the Technology category, followed by Furniture and Office Supplies. 
>>* The East region has the highest average sales overall, followed by the West and Central regions.

>IInd Plot 
>>* The second plot indicates that the highest average profit comes from the Technology category, followed by Office Supplies and Furniture.
>>* The West region has the highest average profit overall, followed by the East and Central regions.



```python
# Group the data by category and sub-category and get the count
csc = data.groupby(['Category', 'Sub-Category']).size()
print(csc)

# Reshape the data into a pivot table
csc = csc.unstack(level=0)


# Plot the stacked bar chart
colors = ['blue', 'green', 'red']
csc.plot(kind='bar', stacked=True, figsize=(20, 10), color=colors)
plt.title('Sub-Category Count by Category')
plt.xlabel('Sub-Category')
plt.ylabel('Count')
plt.legend(title='Category', loc='upper left')
plt.show()
```

#### Insights-
>* The "Office Supplies" category has the most sub-categories with 9, followed by "Furniture" with 4 and "Technology" with 4.
>* The sub-category with the most items is "Binders" with 1522, followed by "Paper" with 1359 and "Storage" with 846.
>* The "Furniture" category has a relatively low number of items in comparison to the other two categories.
>* The "Technology" category has the least number of sub-categories but the items in these sub-categories have relatively high counts.

```python
corr = data.corr()
print(corr)

# Plot the correlation matrix using a heatmap
sns.heatmap(corr, annot=True, cmap='BrBG')
plt.show()
```

#### Insights
>* Sales and Profit are moderately positively correlated (0.48), indicating that increasing Sales can lead to increased Profit.
>* Sales is the most important factor affecting Profit, while Quantity and Discount have a smaller influence on Profit.
>* Quantity and Profit have a weak positive correlation (0.07), suggesting that there is a small tendency for Profit to increase as Quantity increases.
>* Discount and Profit have a weak negative correlation (-0.22), meaning that increasing Discount can cause a slight decrease in Profit.


```python
# Group the data by Category and calculate total sales and profit
category_sales_profit = data.groupby('Category')[['Sales', 'Profit']].sum()

# Create a bar plot to show the total sales and profit for each category
fig, ax = plt.subplots(figsize=(10, 6))
category_sales_profit.plot(kind='bar', ax=ax)
ax.set_title('Total Sales and Profit by Category')
ax.set_xlabel('Category')
ax.set_ylabel('Amount')

plt.show()
```

#### Insights-
>* Technology has Highest Sales and Profit 
>* Priofit margin for all seems moderate but in case of furniture category its very less.
>* Indicates Comapany may need to re-price or improving products and try to understand customer needs in deep related to this Category. 


```python
# Calculate total sales and profit by customer segment
segment_sales_profit = data.groupby('Segment')[['Sales', 'Profit']].sum()
print(segment_sales_profit)
# Create an area plot using stackplot
fig, ax = plt.subplots(figsize=(10, 6))
ax.stackplot(segment_sales_profit.index, segment_sales_profit['Sales'], segment_sales_profit['Profit'], labels=['Sales', 'Profit'])
ax.legend(loc='upper left')
ax.set_title('Customer Segment Sales and Profit')
ax.set_xlabel('Customer Segment')
ax.set_ylabel('Amount')
plt.show()
```

#### Insights-
>*  Consumer segment has the highest sales and profit, followed by the Corporate segment, and the Home Office segment has the lowest sales and profit.
>* It means we can target the Home Office segment for improvement and coming up with strategies to increase sales and profit for that segment.

```python

# Group the data by state and calculate the total sales
state_sales = data.groupby('State')['Sales'].sum().reset_index()
#print(state_sales)

# Sort the data by sales in descending order
state_sales = state_sales.sort_values(by='Sales', ascending=False)

# Create a bar chart to visualize the sales by state
plt.figure(figsize=(14,8))
plt.bar(x=state_sales['State'], height=state_sales['Sales'])
plt.xticks(rotation=90)
plt.xlabel('State')
plt.ylabel('Total Sales')
plt.title('Total Sales by State')
plt.show()
```

#### Insights-
>* California has the highest total sales, followed by New York and Texas.
>* The lowest sales are in North Dakota, West Virginia, and Maine.
>* The analysis of sales figures by state can help businesses optimize their sales strategies to increase revenue and profitability.

```python
# Calculate total profit by state
state_profit = data.groupby('State')['Profit'].sum().sort_values(ascending=False)
# print(state_profit)
# Create a bar plot
fig, ax = plt.subplots(figsize=(15,8))
sns.barplot(x=state_profit.index, y=state_profit.values)
ax.set_title('Total Profit by State')
ax.set_xlabel('State')
ax.set_ylabel('Profit')
plt.xticks(rotation=90)
plt.show()

```

Insights-
>* California, New York, and Washington are the top three states with the highest profit. On the other hand, Texas, Ohio, and Pennsylvania are the three states with the lowest profit.



```python
# Calculate profit margin with sub-category
data['Profit Margin'] = (data['Profit']/data['Sales'])*100
subcat_profit_margin = data.groupby('Sub-Category')['Profit Margin'].mean().sort_values(ascending=False).to_frame().reset_index()
print(subcat_profit_margin)
# Create a horizontal bar plot using Seaborn
fig, ax = plt.subplots(figsize=(10,8))
sns.barplot(x='Profit Margin', y='Sub-Category', data=subcat_profit_margin, ax=ax)
ax.set_title('Profit Margin by Sub-Category')
ax.set_xlabel('Profit Margin (%)')
ax.set_ylabel('Sub-Category')
plt.show()
```

#### Insight 
>* The sub-categories with the highest profit margin are Labels, Paper, and Envelopes, followed by Copiers and Fasteners.
>* The sub-categories with negative profit margin are Binders, Appliances, Tables, Bookcases, and Machines.
